Visualization

```python
# 1. CLASS DISTRIBUTION PLOT

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="CLASS_LABEL",data= df)
plt.title("Phishing vs Legitimate Website Count")
plt.xlabel("Class Label")
plt.ylabel("Frequenccy")
plt.show()
```

OBSERVATION: THE ABOVE GRAPH SHOW THAT THE DATA ARE BALANCED

```python
# 2. FEATURE DISTRIBUTION(UNIVARIATE ANALYSIS)
features_to_plot= ['UrlLength', 'NumDots', 'NumDash', 'NumSensitiveWords']

for feature in features_to_plot:
    plt.figure(figsize=(6,4))
    sns.histplot(df[feature], kde= True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# 3. Correlation Heatmap
plt.figure(figsize=(14,8))
corr= df.corr()
sns.heatmap(corr,cmap='coolwarm', annot= False)
```

```python
# 4. Boxplots by Class

for feature in features_to_plot:
    plt.figure(figsize=(6,4))
    sns.boxplot(x= 'CLASS_LABEL', y= feature, data= df)
    plt.title(f'{feature} by Class Label')
    plt.show()
```
